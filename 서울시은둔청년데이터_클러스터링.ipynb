# 주성분 갯수 다시 정하기

```python
# 주성분 개수를 결정할 때 누적 설명 분산 비율 확인
import numpy as np
pca = PCA().fit(data_c_scaled)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# 원하는 누적 설명 분산 비율을 설정하고 해당되는 주성분 개수 확인
desired_cumulative_variance = 0.95
num_components = np.argmax(cumulative_explained_variance >= desired_cumulative_variance) + 1

print(f"Desired cumulative explained variance: {desired_cumulative_variance}")
print(f"Number of components selected: {num_components}")
```
